```python
import calml
import dummy
```

```python
df = calml.get_hour_features('2014-01-01 00:00:00', '2018-03-15 00:00:00')
```

```python
dummies = dummy.Dummies()
dummies.fit(df, ['year', 'season', 'month', 'monthday', 'weekday', 'hour'])
dm_df = dummies.replace(df)
dm_df.head()
```

```python
dm_df.to_csv('data/calendar.csv')
```
